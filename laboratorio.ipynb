```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

Problema 1

```python
def activar_canal_color(imagen, color: int):
    bgr_imagen = cv2.cvtColor(imagen, cv2.COLOR_RGB2BGR)
    canal_activado = np.zeros_like(bgr_imagen)
    
    if color == 1: # Canal Azul
        canal_activado[:, :, 0] = bgr_imagen[:, :, 0]
    elif color == 2:  # Canal Verde
        canal_activado[:, :, 1] = bgr_imagen[:, :, 1]
    elif color == 3:  # Canal Rojo
        canal_activado[:, :, 2] = bgr_imagen[:, :, 2]
    elif color == 10:  # Canal Rojo y verde
        canal_activado[:, :, 1:3] = bgr_imagen[:, :, 1:3]
    elif color == 20:  # Canal Verde y azul
        canal_activado[:, :, 0] = bgr_imagen[:, :, 0]
        canal_activado[:, :, 1] = bgr_imagen[:, :, 1]
    elif color == 30:  # Canal Azul y rojo
        canal_activado[:, :, 0] = bgr_imagen[:, :, 0]
        canal_activado[:, :, 2] = bgr_imagen[:, :, 2]
    
    return cv2.cvtColor(canal_activado, cv2.COLOR_BGR2RGB)
```

```python
img = cv2.imread('./prueba1.jpg')
```

```python
img_azul = activar_canal_color(img, 1)
plt.imshow(img_azul)
plt.show()
```

Problema 2

```python
def construir_imagen_color(img_gris1, img_gris2, img_gris3):
    imagen_3d = np.dstack((img_gris1, img_gris2, img_gris3))
    
    # Convierte la imagen a la representación de color RGB
    return cv2.cvtColor(imagen_3d, cv2.COLOR_BGR2RGB)
```

```python
img_gris_azul = cv2.imread('./imagen1/imagen1_salida_gray_azul.jpg', 0)
img_gris_verde = cv2.imread('./imagen1/imagen1_salida_gray_verde.jpg', 0)
img_gris_rojo = cv2.imread('./imagen1/imagen1_salida_gray_rojo.jpg', 0)
```

```python
imagen_color = construir_imagen_color(img_gris_azul, img_gris_verde, img_gris_rojo)
plt.imshow(imagen_color)
plt.show()
```

```python
img_gris_azul = cv2.imread('./imagen3/perro_salida_gray_azul.jpg', 0)
img_gris_verde = cv2.imread('./imagen3/perro_salida_gray_verde.jpg', 0)
img_gris_rojo = cv2.imread('./imagen3/perro_salida_gray_rojo.jpg', 0)
```

```python
imagen_color = construir_imagen_color(img_gris_azul, img_gris_verde, img_gris_rojo)
plt.imshow(imagen_color)
plt.show()
```

Problema 3

```python
img = cv2.imread('./prueba1.jpg')
```

```python
def crear_escala_grises_3d(imagen):
    bgr_imagen = cv2.cvtColor(imagen, cv2.COLOR_RGB2BGR)
    canal_rojo = bgr_imagen[:, :, 2]
    canal_verde = bgr_imagen[:, :, 1]
    canal_azul = bgr_imagen[:, :, 0]

    # Crea las imágenes en escala de grises para cada canal de color
    img_gris_rojo = cv2.cvtColor(canal_rojo, cv2.COLOR_GRAY2RGB)
    img_gris_verde = cv2.cvtColor(canal_verde, cv2.COLOR_GRAY2RGB)
    img_gris_azul = cv2.cvtColor(canal_azul, cv2.COLOR_GRAY2RGB)

    return img_gris_rojo, img_gris_verde, img_gris_azul
```

```python
img_gris_rojo, img_gris_verde, img_gris_azul = crear_escala_grises_3d(img)
```

```python
cv2.imwrite('./prueba1_gris_rojo.jpg', img_gris_rojo)
plt.imshow(img_gris_rojo)
plt.show()
```

```python
cv2.imwrite('./prueba1_gris_verde.jpg', img_gris_verde)
plt.imshow(img_gris_verde)
plt.show()
```

```python
cv2.imwrite('./prueba1_gris_azul.jpg', img_gris_azul)
plt.imshow(img_gris_azul)
plt.show()
```
